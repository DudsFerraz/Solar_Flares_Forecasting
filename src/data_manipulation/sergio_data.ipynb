```python
import os
import pandas as pd
from dotenv import load_dotenv
```

```python
load_dotenv()

data_path = os.getenv('SERGIO_DATA_PATH')
files_names = ["All_2010_2017.csv","All_2010_2017_slided.csv","All_2010_2017_slided_withouNaN.csv"]
df_names = ["default","slided","no_nan"]
data = {}
```

```python
for file, df_name in zip(files_names,df_names):
    full_path = os.path.join(data_path, file)

    df = pd.read_csv(full_path, index_col="t_rec")
    df.index = pd.to_datetime(df.index, format='%Y.%m.%d_%H:%M:%S_TAI')
    df.index.name = "ds"

    data[df_name] = df
```
